```python
import pandas as pd

df = pd.read_csv("pizzas.csv")
df
```

```python
df.plot(kind="scatter", x="diameter", y="price");
```

```python
from sklearn import linear_model

model = linear_model.LinearRegression()
x = df[["diameter"]]
y = df[["price"]]

model.fit(x, y)
```

```python
print(model.predict([[42]])[0][0])
```
